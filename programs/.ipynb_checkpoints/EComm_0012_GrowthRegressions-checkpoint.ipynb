# Loading SITC 2015 data 
from: https://intl-atlas-downloads.s3.amazonaws.com/index.html

```python
ccpy_filepath = "https://intl-atlas-downloads.s3.amazonaws.com/CCPY/S2_final_{yr}.dta"
cpy_filepath = "https://intl-atlas-downloads.s3.amazonaws.com/CPY/S2_final_cpy_all.dta"
ctyregions_filepath = "https://raw.githubusercontent.com/lukes/ISO-3166-Countries-with-Regional-Codes/master/all/all.csv"
gdp_filepath = "C:\\Users\\agomez\\Dropbox\\Harvard\\LittleProjects\\StochasticPS\\data\\WorldBank_GDPperCapita_1962_to_2015.json"
```

```python
colimporter = 'importer'
colprod = 'commoditycode'
newcol = 'imp_prod'

rowvarstring = 'exporter'
colvarstring = 'commoditycode'
valvarstring = 'export_value'

rcaval = 'rca_ols'
```

```python
ctyregs_df = pd.read_csv(ctyregions_filepath)
ctyregs_df.head()
```

```python
ctyregs_df = ctyregs_df[~ctyregs_df['sub-region'].isnull()]
ctyregs_df.shape
```

```python
###### LOADING THE DATA #######
longdf_cp = pd.read_stata(cpy_filepath)
```

```python
print(longdf_cp.shape)
longdf_cp.head()
```

```python
exp_codes = set(longdf_cp['exporter'])
iso_codes = set(ctyregs_df['alpha-3'])
cty_codes = set(exp_codes) & set(iso_codes)

# Subsetting and cleaning
longdf_cp = longdf_cp[longdf_cp["exporter"].isin(cty_codes) & 
                      (longdf_cp["export_value"] > 0) & 
                      (longdf_cp["commoditycode"] != ".")].drop(["inatlas",
                                                                 "oppval",
                                                                 "oppgain",
                                                                 "distance",
                                                                 "import_value"], 1)

```

```python
# Creating totals
longdf_cp = longdf_cp.merge(longdf_cp.groupby(by = ["year", "exporter"])[["export_value"]].sum(), left_on = ["year", "exporter"], right_index = True, suffixes = ("", " exporter tot"))
longdf_cp = longdf_cp.merge(longdf_cp.groupby(by = ["year", "commoditycode"])[["export_value"]].sum(), left_on = ["year", "commoditycode"], right_index = True, suffixes = ("", " commoditycode tot"))

# Running regression
longdf_cp["constant"] = 1
longdf_cp[["log_export_value", "log_export_value exporter tot", "log_export_value commoditycode tot"]] = np.log(longdf_cp[["export_value", "export_value exporter tot", "export_value commoditycode tot"]])

rca_regression = OLS(longdf_cp["log_export_value"], exog = longdf_cp[["log_export_value exporter tot", "log_export_value commoditycode tot", "constant"]], hasconst = True).fit()
longdf_cp[rcaval] = (rca_regression.resid > 0).astype(int)  
```

```python
longdf_cp.head()
```

# Looping through the years
## 1) Load the data for the year
## 2) Create the Mcp and Mcpi
## 3) Compute the Ccc_cp and Ccc_cpi
## 4) Compute the left and right eigenvectors
## 5) Compute the distance to the center
## 6) Append distance to a long list of countries and years.

```python
dists_allyears_df = longdf_cp[['year', 'exporter', 'population', 'eci']].drop_duplicates(subset=['year', 'exporter'], keep='last')
dists_allyears_df['dist_from_cp'] = 0
dists_allyears_df['dist_from_ccp'] = 0
```

```python
dists_allyears_df.tail()
```

```python
years = np.arange(1962, 2017)
```

```python

for year in years:
    print("Starting the year {}".format(year))
    
    # 1) Load the data cpi for the year
    longdf_ccpy = pd.read_stata(ccpy_filepath.format(yr=year))
    longdf_cpy = longdf_cp[longdf_cp['year']==year][[rowvarstring, colvarstring, rcaval]]
    
    longdf_ccpy = longdf_ccpy[longdf_ccpy[rowvarstring].isin(cty_codes) & 
                      (longdf_ccpy[valvarstring] > 0) & 
                      (longdf_ccpy[colprod] != ".")]
    
    # compute discrete rca
    longdf_ccpy = EComm_0001_complexities.RCA_from_longOLS_insingleyear(longdf_ccpy,
                                                                      dim_vars=[rowvarstring, colimporter, colprod],
                                                                      value_var=valvarstring, 
                                                                      new_var=rcaval,
                                                                      discrete=True)
    longdf_ccpy[newcol] = longdf_ccpy[colimporter]+longdf_ccpy[colprod] 

    print("Ready to start to create the mcps")
    # 2) Create the Mcp and Mcpi
    Mccp_widedf = longdf_ccpy[[rowvarstring, newcol, rcaval]].pivot(index=rowvarstring, 
                                                                   columns=newcol, 
                                                                   values=rcaval).fillna(0.0)    
    # From long to wide format
    Mcp_widedf = longdf_cpy[[rowvarstring, colvarstring, rcaval]].pivot(index=rowvarstring, 
                                                                       columns=colvarstring, 
                                                                       values=rcaval).fillna(0.0)
    
    # 3) Compute the Ccc_cp and Ccc_cpi and 4) Compute the left and right eigenvectors
    #(Mc2c_ccp, Dc_ccp, leftVc_ccp, rightVc_ccp), (Mp2p_ccp, Dp_ccp, leftVp_ccp, rightVp_ccp) = EComm_0001_complexities.ECeigenvecs(Mccp_widedf)
    #(Mc2c_cp, Dc_cp, leftVc_cp, rightVc_cp), (Mp2p_cp, Dp_cp, leftVp_cp, rightVp_cp) = EComm_0001_complexities.ECeigenvecs(Mcp_widedf)
        
    print("Ready to compute the distances")
    # 5) Compute the distance to the center
    dists_ccp = EComm_0001_complexities.distance_to_center(Mccp_widedf, kcomm=5, distvarname = 'dist_from_ccp')
    dists_cp = EComm_0001_complexities.distance_to_center(Mcp_widedf, kcomm=2, distvarname = 'dist_from_cp')
    
    print("Generating the dataframe")
    # 6) Append distance to a long list of countries and years.
    dists_ccp[rowvarstring] = Mccp_widedf.index.values
    dists_cp[rowvarstring] = Mcp_widedf.index.values
    
    dists_ccp['year'] = year
    dists_cp['year'] = year
    
    print("Merging")
    # 7) Merging
    dists_allyears_df = dists_allyears_df.merge(dists_ccp, how='left', on = ["year", "exporter"], suffixes = ("_x", "_y"))
    dists_allyears_df['dist_from_ccp'] = np.max(dists_allyears_df[['dist_from_ccp_x', 'dist_from_ccp_y']], axis=1)
    dists_allyears_df = dists_allyears_df.drop(labels=['dist_from_ccp_x', 'dist_from_ccp_y'], axis=1)
    
    dists_allyears_df = dists_allyears_df.merge(dists_cp, how='left', on = ["year", "exporter"], suffixes = ("_x", "_y"))
    dists_allyears_df['dist_from_cp'] = np.max(dists_allyears_df[['dist_from_cp_x', 'dist_from_cp_y']], axis=1)
    dists_allyears_df = dists_allyears_df.drop(labels=['dist_from_cp_x', 'dist_from_cp_y'], axis=1)
    
    # take the values that of interest from the clashes

    print("end of cycle")
    print("")

    
```

### Which communities to use?

```python
communitycolumn = 'region'

# left-eigenvalue data frame
Vc_df = pd.DataFrame(Vc, index=Mc2c.index)
Vc_df['Community'] = [ctyregs_df[ctyregs_df['alpha-3']==cname][communitycolumn].values[0] for cname in Mc2c.index.values]

# right-eigenvalue data frame
rightVc_df = pd.DataFrame(rightVc, index=Mc2c.index)
rightVc_df['Community'] = [ctyregs_df[ctyregs_df['alpha-3']==cname][communitycolumn].values[0] for cname in Mc2c.index.values]

print(Vc_df.head())
print(rightVc_df.head())
```
