```python
###### LOADING THE DATA #######
longdf_cpy = pd.read_stata(cpy_filepath)
print(longdf_cpy.shape)
longdf_cpy.head()
```

```python
exp_codes = set(longdf_cpy['exporter'])
iso_codes = set(ctyregs_df['alpha-3'])
cty_codes = set(exp_codes) & set(iso_codes)

# Subsetting and cleaning
longdf_cpy = longdf_cpy[longdf_cpy["exporter"].isin(cty_codes) & 
                      (longdf_cpy["export_value"] > 0) & 
                      (longdf_cpy["commoditycode"] != ".")].drop(["inatlas",
                                                                 "oppval",
                                                                 "oppgain",
                                                                 "distance",
                                                                 "import_value"], 1)

```

```python
# Creating totals
longdf_cpy = longdf_cpy.merge(longdf_cpy.groupby(by = ["year", "exporter"])[["export_value"]].sum(), left_on = ["year", "exporter"], right_index = True, suffixes = ("", " exporter tot"))
longdf_cpy = longdf_cpy.merge(longdf_cpy.groupby(by = ["year", "commoditycode"])[["export_value"]].sum(), left_on = ["year", "commoditycode"], right_index = True, suffixes = ("", " commoditycode tot"))

# Running regression
longdf_cpy["constant"] = 1
longdf_cpy[["log_export_value", "log_export_value exporter tot", "log_export_value commoditycode tot"]] = np.log(longdf_cpy[["export_value", "export_value exporter tot", "export_value commoditycode tot"]])

rca_regression = OLS(longdf_cpy["log_export_value"], exog = longdf_cpy[["log_export_value exporter tot", "log_export_value commoditycode tot", "constant"]], hasconst = True).fit()
longdf_cpy[rcaval] = (rca_regression.resid > 0).astype(int)  
```

## Converting long to wide

```python
colimporter = 'importer'
colprod = 'commoditycode'
newcol = 'imp_prod'

rowvarstring = 'ctyyear'
colvarstring = 'commoditycode'
#valvarstring = 'export_value'
valvarstring = 'mcp'

rcaval = 'rca_ols'
```

```python
# ---------------------
# From long to wide format
#Mcp_widedf = longdf_cpy[(longdf_cpy['year'].astype(int)%10==0)][[rowvarstring, colvarstring, valvarstring]].pivot(index=rowvarstring, 
Mcp_widedf = longdf_cpy[[rowvarstring, colvarstring, valvarstring]].pivot(index=rowvarstring, 
                                                                       columns=colvarstring, 
                                                                       values=valvarstring).fillna(0.0)
```

```python
# Getting rid of the rows/columns of pure zeros
Mcp_widedf = Mcp_widedf.loc[:, (Mcp_widedf != 0).any(axis=0)]
Mcp_widedf = Mcp_widedf.loc[(Mcp_widedf != 0).any(axis=1), :]
Mcp_mat = np.array(Mcp_widedf)
print(Mcp_widedf.shape)
```
