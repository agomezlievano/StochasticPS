## Preparation

```python
# Creating totals
rowtotalvar = valtotalstring + "_{}_tot".format(rowvarstring)
coltotalvar = valtotalstring + "_{}_tot".format(colvarstring)
totaltotalvar = valtotalstring + "_year_tot"

longdf[rowtotalvar] = longdf.groupby(by = ["year", rowvarstring])[valtotalstring].transform('sum')
longdf[coltotalvar] = longdf.groupby(by = ["year", colvarstring])[valtotalstring].transform('sum')
longdf[totaltotalvar] = longdf.groupby(by = ["year"])[valtotalstring].transform('sum')

longdf = longdf.loc[( (longdf[rowtotalvar]>0) & (longdf[coltotalvar]>0) )]
```

```python
# Creating logarithms
longdf["constant"] = 1
longdf["log_" + valtotalstring] = np.log(1 + longdf[valtotalstring].values)
longdf[["log_" + rowtotalvar, "log_" + coltotalvar, "log_" + totaltotalvar]] = np.log(longdf[[rowtotalvar, coltotalvar, totaltotalvar]])
```

```python
Xmat = longdf[["log_" + rowtotalvar, "log_" + coltotalvar, "log_" + totaltotalvar]].values
ymat = longdf["log_" + valtotalstring].values
reg = LinearRegression().fit(Xmat, ymat)
```

```python
# creating the residuals of a regression
rcaval = "rca_ols"
mcpval = "mcp_ols"
longdf[rcaval] = ymat - reg.predict(Xmat)  
longdf[mcpval] = (longdf[rcaval] > 0).astype(int)  
```

```python
def plot_moving_country(country_code, finalyear, pca, data, axis, 
                        dopivot = True,
                        color='r', linewidth=2, head_length=.7, head_width = .7, **kwargs):
    """
    Plot the trajectory of a certain country. 
    """
    #subdf = data_pivot_table.query('year <= @finalyear')
    #cty = pca.transform(subdf.loc[(country_code,slice(None)), 'RCA_ave_transformed'].fillna(value=0).values)
    
    if dopivot:
        cty = pca.transform(
            (data.pivot_table(index=["year", rowvarstring],
                              columns=colvarstring,
                              values=rcaval).fillna(value=0).query("year <= @finalyear").loc[(slice(None),country_code),:].values))
    else:
        cty = pca.transform((data.query("year <= @finalyear").loc[(slice(None),country_code),:].values))
    
    numyears = cty.shape[0]

    # Interpolation
    mytck,myu = interpolate.splprep([cty[:,0],cty[:,1]], s=4)
    xnew2,ynew2 = interpolate.splev(np.linspace(0,1,numyears*5),mytck)


    NPOINTS = len(xnew2)
    xi = 30
    for i in range(NPOINTS-1):
        axis.plot(xnew2[i:i+2],ynew2[i:i+2],
              alpha=np.exp((float(i)-NPOINTS+2.0)/float(xi)), c=color, linewidth=linewidth,
              label=country_code, **kwargs)   
    axis.plot(xnew2[-1],ynew2[-1], c=color, marker='.', markersize=10)

    # write a label of the country
    plt.text(cty[-1,0], cty[-1,1], country_code, fontsize = 14, bbox={'facecolor':color, 'alpha':0.3, 'pad':3})
    
```
